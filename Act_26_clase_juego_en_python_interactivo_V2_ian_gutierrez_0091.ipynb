```python
class Juego:
    def __init__(self, nombre, plataforma, genero):
        self.nombre = nombre
        self.plataforma = plataforma
        self.genero = genero

    def leerdatos(self):
        self.nombre = input("ingrese el nombre del juego: ")
        self.plataforma = input("ingrese la plataforma del juego: ")
        self.genero = input("ingrese el genero del juego: ")
        return self.nombre, self.plataforma, self.genero

    def mostrardatos(self):
        print("nombre: ", self.nombre)
        print("plataforma: ", self.plataforma)
        print("genero: ", self.genero)
        return self.nombre, self.plataforma, self.genero
#fin de clase Juego

print("juego numero 1")
juego1 = Juego("", "", "") # creando objeto juego 1 con valores iniciales
juego1.leerdatos()
juego1.mostrardatos()

print("-*-*-*-*-*-*-*-*-*-*-*")

print("juego numero 2")
juego2 = Juego("", "", "") # creando objeto juego 2 con valores iniciales
juego2.leerdatos()
juego2.mostrardatos()

print("-*-*-*-*-*-*-*-*-*-*-*")
print("sofware desarrollado por ian gutierrez NL = 22")
```
